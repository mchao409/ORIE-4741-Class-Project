```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv("weather.csv")
```

```python
weather.head()
```

```python
sampled = weather.sample(frac=1)
```

```python
temps = sampled["mean_temperature_f"]
temp_counts = temps.value_counts()
```

```python
i = temp_counts.index
v = temp_counts.values
```

```python
plt.bar(i, v)
plt.xlabel("Temperature")
plt.ylabel("Number of Riders")
```
