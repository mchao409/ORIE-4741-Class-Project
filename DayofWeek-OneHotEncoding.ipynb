```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv("needed_features.csv")
```

```python
features['weekday'] = features['weekday'].astype(str)
```

```python
sunday = [1, 0, 0, 0, 0, 0, 0]
monday = [0, 1, 0, 0, 0, 0, 0]
tuesday = [0, 0, 1, 0, 0, 0, 0]
wednesday = [0, 0, 0, 1, 0, 0, 0]
thursday = [0, 0, 0, 0, 1, 0, 0]
friday = [0, 0, 0, 0, 0, 1, 0]
saturday = [0, 0, 0, 0, 0, 0, 1]
```

```python
mapping_dict={'Sunday': sunday,'Monday': monday,'Tuesday': tuesday,'Wednesday': wednesday,'Thursday': thursday,'Friday': friday,'Saturday': saturday}
```

```python
features['weekday']=features['weekday'].map(mapping_dict)
```
